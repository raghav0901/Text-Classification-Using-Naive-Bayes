```python

#BUILDING A DICTIONARY of DIFFERENT WORDS APPEARING IN DOCUMENTS ALONG WITH WORD COUNTS:
all_classes = os.listdir("20_newsgroups")
print(all_classes)
# creating the word dictionary
word_dictionary = {}

"""
DICTIONARY :
    Keys   : words
    Values : count of word
"""

# using count for checking progress during runtime
count = 1

# Iterating through all classes
for class_ in all_classes:
    # storing the all documents in a list for a particular class
    all_docs = os.listdir("20_newsgroups"+"/"+class_)
    
    # traversing through all documents
    for docs in all_docs:
        
        # getting the base path which is common for all documents of a particular class
        path = "20_newsgroups"+"/"+class_+"/"+docs
        
        # opening the particular document as file object
        with open(path, encoding = "latin-1") as file_obj:
            
            # reading the document data
            file_data = file_obj.read()
            
            # stroing each words in a list , used re to remove the whitespaces 
            tokens = re.compile('\w+').findall(file_data)
            
            # traversing through all words and updating the value in dictionary if already exist otherwise will add to it
            for word in tokens:
                
                # checking if the word in a stop_word or not, is aphbetic and the lenth should be > 3 to avoid words like i, me etc
                if (word.lower() not in stop_words) and (word.lower().isalpha()) and (len(word.lower())>3):
                    
                    # getting the value of word from dictionary and increasing its count by 1 if word does not exists in dictionart then adding it with count 1
                    word_dictionary[word.lower()] = word_dictionary.get(word.lower(),0) + 1 
    # using print statement to check the progress during runtime
    print(count, "Class done")
    count+=1
    
'''Sorting the dictionary: '''     
sort_orders = sorted(word_dictionary.items(), key=lambda x: x[1], reverse=True)

''' Choosing the top 1000 words as the features of the vocabulary'''
newpass=[x[0] for x in sort_orders[:1000]]

''' Building a new dictionary that will be used to build a dataframe with the vocabulary features as the keys '''
newdict={}
print(newpass)

for x in newpass:
    newdict[x]=[]
```

```python
newdict['Target']=[]

''' We repeat the process of going through each document again word by word, storing the frequency of words in the vocabulary'''

#storing the all the classe names in a list from the given folder
all_classes = os.listdir("20_newsgroups")

# Iterating through all classes
for class_ in all_classes:
    # storing the all documents in a list for a particular class
    all_docs = os.listdir("20_newsgroups"+"/"+class_)
    
    # traversing through all documents
    for docs in all_docs:
        
        # getting the base path which is common for all documents of a particular class
        path = "20_newsgroups"+"/"+class_+"/"+docs
        
        # opening the particular document as file object
        with open(path, encoding = "latin-1") as file_obj:
            newdict['Target'].append(class_)
            for x in newpass:
                newdict[x].append(0)
            
            # reading the document data
            file_data = file_obj.read()
            
            # stroing each words in a list , used re to remove the whitespaces 
            tokens = re.compile('\w+').findall(file_data)
            # traversing through all words and updating the value in dictionary if already exist otherwise will add to it
            for word in tokens:
                # checking if the word in a stop_word or not, is aphbetic and the lenth should be > 3 to avoid words like i, me etc
                if (word.lower() not in stop_words) and (word.lower().isalpha()) and (len(word.lower())>3) and (word.lower() in newpass):
                     newdict[word.lower()][-1]=newdict[word.lower()][-1]+1
```

```python
''' Building a numpy 2-d array out of the X-dataframe'''
import pandas as pd
df = pd.DataFrame(newdict)
X=df.iloc[:,:-1].copy().to_numpy()
print(X)
''' Building a numpy 1-d array out of the Y-series(Target)'''
Y=df.iloc[:,-1].copy().to_numpy()
print(Y)
```

```python
# Getting ready to use the in built Naive baiyes classifier:
## Train-Test splitting first:
from sklearn import model_selection
# Splitting the data into train - test, using random state to be 0 
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,test_size=0.25,random_state=0)
print(Y_test)
```

```python
'''Using the in built MultinomialNaive Baiyes'''

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
# creating a classifier object
clf = MultinomialNB()

# fitting the classifier object to training data
clf.fit(X_train, Y_train)

# testing the classifier and storing the predicted Y_pred classes
Y_pred = clf.predict(X_test)
# Generating the classification report and printing the confusion matrix
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python

# Making a dictionary which will store the count of vocabulary words,,according to the different Y-target class values
grandy={}

# Declaring other helpful variables
df=None
sery=None
df1=None

## Writing my own naive baiyes::
# function used for fitting training data
def fit(X_train,Y_train):
    '''
    FUNCTION :
            fit : 
                Arguments : Training Data (X_train and Y_train)
        returns   : 
                A DICTIONARY :
                    Keys   : class_names and total count of all features
                    Values : DICTIONARY
                         Keys   : Words and total count of word of current_class
                         Values : number of times word appears in a sample of current_class 
    '''
    global df
    global df1
    global sery
    df = pd.DataFrame(data=X_train,columns=newpass)
    sery = pd.Series(Y_train)
    df1= pd.concat([df, sery], axis=1)
    unique_ys=sery.value_counts().index
    for x in unique_ys:
        grandy[x]=((df[df1[0]==x]).sum().to_dict())
        grandy[x]['Total']=((df[df1[0]==x]).sum()).sum()
fit(X_train,Y_train)
   
```

```python
## Finding probability of test data belonging to a specific class:
import numpy as np
def prob_specific(dic,test,class_):
    log_prob=np.log(len(df[df1[0]==class_].index)/len(df.index))
    test_series=pd.Series(test)
    test_series.index=newpass
    for x in newpass:
        if test_series[x]==0:
            continue
        else:
            ## adding the log of probabilitiesusing laplace correction:
            log_prob+=np.log((dic[class_][x]+1)/(dic[class_]['Total']+len(newpass)))
    return log_prob
```

```python
### Finding predicted class :
def predicted_class(dic,test):
    unique_ys=sery.value_counts().index
    curr_pred_clas=None
    curr_prob=-100000
    for x in unique_ys:
        if curr_prob<prob_specific(dic,test,x):
            curr_pred_clas=x
            curr_prob=prob_specific(dic,test,x)
    return curr_pred_clas
    

## Function used for predicting class of every test data point:
def predict_every(dic,Xtest):
    ypred=[]
    counter=0
    for x in Xtest:
        r=predicted_class(dic,x)
        ypred.append(r)
        counter+=1
        print(counter/len(Xtest)*100)
    return ypred

Y_pred=predict_every(grandy,X_test)

# generating the classification report and printing the confusion matrix
print(classification_report(Y_test,Y_pred))


    
```
